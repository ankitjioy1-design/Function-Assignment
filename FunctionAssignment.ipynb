4 .The return statement in a Python function serves two main purposes: it ends the execution of the function and sends a value back to the caller. When Python encounters a return statement during function execution, the function immediately stops running, and the expression (if any) following return is evaluated and passed back as the result.

5 . What are iterators in Python and how do they differ from iterables?

5 . ***Iterable***

An iterable is any Python object capable of returning its members one at a time, typically used in loops like for. Common examples include lists, tuples, strings, dictionaries, and sets. Technically, an iterable implements the __iter__() method, which returns an iterator object for traversing its content.

***Iterator***

An iterator is a specific object used to perform iteration over an iterable. It implements two methods:

*   __iter__(): Returns itself as an iterator.
*   __next__(): Returns the next item. Raises StopIteration when it reaches the end.

The main difference is that iterables are objects you can loop over, but iterators are objects that do the actual iteration and remember where they are in the process. An iterable can provide a fresh iterator every time, while an iterator produces each item one after another and gets exhausted once all items are retrieved.

 7 . What are the advantages of using generators over regular functions?

 7 .The main advantages of using generators over regular functions in Python are:



*   ***Memory Efficiency***

Generators produce values one at a time and do not store the entire sequence in memory. This makes them highly efficient when working with large datasets or potentially infinite sequences, unlike regular functions that return a complete list in memory.

*   ***Lazy Evaluation***

Generators yield values only when requested, allowing computations to be paused and resumed. This lazy evaluation improves performance as values are generated on-the-fly rather than upfront.​


​

*   ***State Retention***

Generators remember their execution state between yields, enabling complex iterated computations without recomputing or managing external state manually.​
*   ***Simpler Syntax for Iteration***

Using yield makes it concise and cleaner to write iterators, avoiding the overhead of managing custom iterator classes or complete list creation.



*   ***Pipeline Processing***

Generators can be chained to form data pipelines that process data in stages efficiently without intermediate data storage.

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 10 .  map()


Purpose: Applies a given function to each item in an iterable (like list, tuple) and returns an iterator that yields transformed items.

Use case: When you want to modify or transform all elements in a collection.

Example:

python
numbers = [1, 2, 3]

squared = list(map(lambda x: x * x, numbers))

filter()

Purpose: Uses a boolean function to select items from an iterable, returning only those for which the function returns True.

Use case: When you want to filter out certain elements based on a condition.

Example:

python
numbers = [1, 2, 3, 4]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

reduce()

Purpose: Applies a binary function cumulatively to the items in an iterable, reducing it to a single value.

Use case: When you want to combine all elements into one result, such as sum or product.

Example:

from functools import reduce

numbers = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, numbers)

```python
 """5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms. """

class FibonacciIterator:
    def __init__(self, n):
        self.n = n          # Total terms to generate
        self.count = 0      # Current count
        self.a = 0          # First Fibonacci number
        self.b = 1          # Second Fibonacci number

    def __iter__(self):
        return self         # The iterator object itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            fib = self.a + self.b
            self.a, self.b = self.b, fib
            self.count += 1
            return fib

# Example usage:
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)
```

```python
""" 7. Implement a generator function that reads a file line by line and
yields each line as a string."""

def read_file_generator(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Yield each line without the newline character

# Example usage:
for line in read_file_generator('example.txt'):
    print(line)
```

```python
"""10. Create a Python program that uses `filter()` to remove all the vowels from
 a given string."""
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    # filter out characters that are vowels
    filtered_chars = filter(lambda char: char not in vowels, input_string)
    # join filtered characters back into a string
    return ''.join(filtered_chars)

# Example usage:
text = "Hello, how are you?"
result = remove_vowels(text)
print(result)  # Output: Hll, hw r y?
```
